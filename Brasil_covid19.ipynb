```
file = '/content/drive/My Drive/DataScience/brazil_covid19.csv'
df = pd.read_csv(file, sep=',', parse_dates=['date'])
```

```
import warnings
warnings.filterwarnings('ignore')
```

```
df.info()
```

```
# Casos recuperados = Casos confirmados - mortes 
df['recuperados'] = df['cases'] - df['deaths'] 
```

```
# Preenchendo missing values 
df[['cases', 'deaths', 'recuperados']] = df[['cases', 'deaths', 'recuperados']].fillna(0)
```

```
# imprimindo 5 primeiras linhas
df.head()
```

```
df.date.describe()
```

```
# Obtem o numero de casos confirmados, mortes e recuperados por data e por região.
df_agrupado = df.groupby(['date', 'state'])['cases', 'deaths', 'recuperados'].sum().reset_index()
```

```
# Ordena o dataframe por casos confirmados
df_agrupado.sort_values(by='cases', ascending=False)
```

```
# Obtem o número de casos confirmados, mortes e recuperados agrupando por região.
df_group_estado = df.groupby('state')['cases', 'deaths', 'recuperados'].sum().reset_index()
```

```
# ordena por estado com mais casos confirmados
df_group_estado.sort_values(by='cases', ascending=False)
```

```
# Agrupa quantidade de casos, mortes e recuperados por data
temp = df.groupby('date')['cases', 'deaths','recuperados'].sum().reset_index()
```

```
# Remodela o dataframe com variável e valor para ter quantidades de casos, mortes e recuperados
temp = temp.melt(id_vars="date", value_vars=['cases', 'deaths', 'recuperados'],
                 var_name='Case',value_name='Count')
```

```
temp.head(20)
```

```
# Cores
casos = '#21bf73'
mortes = '#ff2e63'
recuperados = '#fe9801'
```

```
import plotly.express as px
fig = px.area(temp, 
              x="date", 
              y="Count", 
              color='Case', 
              height=600,
              title='Casos somados ao longo do tempo',
              color_discrete_sequence = [casos,mortes,recuperados])
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```

```
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
import plotly.express as px
```

```
# filtra os dados considerando o último dia da base de dados
completo = df[df['date'] == max(df['date'])]
```

```
# imprime as 5 primeiras linhas
completo.head()
```

```
fig = px.line(df_agrupado,
              x="date",
              y="cases",
              color='state',
              height=600,
              title='Casos Confirmados',
              color_discrete_sequence = px.colors.qualitative.Dark2 )
fig.show()

fig = px.line(df_agrupado,
              x="date",
              y="deaths",
              color='state',
              height=600,
              title='Mortes Confirmadas',
              color_discrete_sequence = px.colors.qualitative.Dark2)
fig.show()
```

```
# Plota painel                        
fig = px.treemap(completo.sort_values(by='cases', ascending=False).reset_index(drop=True), 
                 path=["state", "region"],
                 values="cases",
                 height=600,
                 title='Número de Casos Confirmados',
                 color_discrete_sequence = px.colors.qualitative.Dark2)
fig.data[0].textinfo = 'label+text+value'
fig.show()
```

```
# Plota Painel
fig = px.treemap(completo.sort_values(by='deaths', ascending=False).reset_index(drop=True), 
                 path=["state", "region"],
                 values="deaths",
                 height=600,
                 title='Número de Mortes Confirmadas',
                 color_discrete_sequence = px.colors.qualitative.Dark2)
fig.data[0].textinfo = 'label+text+value'
fig.show()
```
